```python
def step_gradient(x, y, learning_rate, m):
    m_slope = np.zeros(x.shape[1])
    M = len(x)
    for i in range(M):
        X= x[i, :]
#         print(X.shape, m.shape)
        Y= y[i]
        for j in range(len(X)):
            m_slope[j] += (-2/M)* (Y - sum( m * X))*X[j]
    new_m = m - learning_rate* m_slope   
    return new_m
```

```python
def cost(x, y, m):
    total_cost = 0
    M = len(x)
    for i in range(M):
        Y= y[i]
        X= x[i]
        total_cost += (1/M)*((Y - sum(m*X))**2)
    return total_cost
```
