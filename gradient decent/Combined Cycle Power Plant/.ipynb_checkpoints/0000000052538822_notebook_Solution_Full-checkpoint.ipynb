```python
import numpy as np
import pandas as pd
bdata=pd.read_csv("train.csv")
bdata
```

```python
x_train_boston_data=bdata.iloc[:,0:13]
x_train_boston_data
```

```python
y_train_boston_data=bdata.iloc[:,-1]
y_train_boston_data
```

```python

from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor()
model.fit(x_train_boston_data,y_train_boston_data)
boston_data1=np.loadtxt("test.csv",delimiter = ',')
y_predict=model.predict(boston_data1)
np.savetxt("boston_predict.csv", y_predict,fmt='%.5f')
y_predict
```
